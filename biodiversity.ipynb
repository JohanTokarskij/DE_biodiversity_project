```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
observations = pd.read_csv('observations.csv')
species = pd.read_csv('species_info.csv')

observations.head()
```

```python
# What is the distribution of conservation_status for animals?

plt.figure(figsize=(10, 6))
sns.countplot(data=species, x='conservation_status')
plt.title('Conservation Status Distribution')
plt.xlabel('Conservation Status')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
plt.clf()
```

```python
# Are certain types of species more likely to be endangered?

plt.figure(figsize=(10, 6))
sns.countplot(data=species, x='category', hue='conservation_status')
plt.title('Conservation Status by Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.legend(title='Conservation Status', loc='upper right')
plt.xticks(rotation=45)
plt.show()
```

```python
# Which species were spotted the most at each park?

most_spotted_species = observations.groupby('park_name')['observations'].idxmax()
most_spotted_df = observations.loc[most_spotted_species, ['scientific_name', 'park_name', 'observations']]

most_spotted_df

```
